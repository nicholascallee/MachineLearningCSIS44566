## Part 1 - Read your data frame.
The CSV file is "fifa_data.csv"

```python
# Your code here
import pandas as pd
data = pd.read_csv("fifa_data.csv")

print(data.info())

data.drop(labels=["Loaned From", "Club", "Joined"], axis='columns', inplace=True)
data.dropna(axis='index', how='any', inplace=True)
print(data)

data.hist(bins=20)
print(data.head(n=10))
data.corr()

print(data.keys())
print(data.info())
print()
print(data["Nationality"].value_counts())

from pandas.plotting import scatter_matrix


def euro_converter(string):
    unit = string[0:1]
    value = string[1:len(string)-1]
    multiplier = string[-1]
    if multiplier =="0": return 0
    
    # print(unit, value, multiplier)
    value = float(value)
    if multiplier=="M": value *= 1000000
    if multiplier=="K": value *= 1000
    return value

data["Numeric Wage"] = data["Wage"].map(euro_converter)

compare = ["Overall", "Potential", "HeadingAccuracy", "Age", "Skill Moves", "Numeric Wage"]
scatter_matrix(data[compare])

```
